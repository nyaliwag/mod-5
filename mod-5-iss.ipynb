```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
import time
import numpy as np
%matplotlib inline

def get_space_station_location():
    r = requests.get(url='http://api.open-notify.org/iss-now.json')
    space_station_location = (r.json())

x_list=[]
for i in range(101):
    r = requests.get(url='http://api.open-notify.org/iss-now.json')
    space_station_location = (r.json())
    space_station_latitude = float(space_station_location['iss_position']['latitude'])
    x_list.append(space_station_latitude)
    time.sleep(10)
    
y_list=[]
for i in range(101):
    r = requests.get(url='http://api.open-notify.org/iss-now.json')
    space_station_location = (r.json())
    space_station_longitude = float(space_station_location['iss_position']['longitude'])
    y_list.append(space_station_longitude)
    time.sleep(10)

    
t_list=[]
for i in range(101):
    r = requests.get(url='http://api.open-notify.org/iss-now.json')
    space_station_location = (r.json())
    timestamp = float(space_station_location["timestamp"])
    t_list.append(timestamp)
    time.sleep(10)

    
    
m_list=list(zip(x_list, y_list, t_list))
print(m_list)
```

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
fig.set_size_inches(20, 19)
ax = plt.axes(projection='3d')

zdata = [t_list]
xdata = [x_list]
ydata = [y_list]
ax.scatter3D(xdata, ydata, zdata, cmap='Blues')
```
